## Metricas de Desempeño (KPI) generales de sistemas de recomendación:

- **CTR (Click-Through Rate)**:
  - Mide la cantidad de clics en los ítems recomendados.
  - Representa la relevancia percibida de las recomendaciones.

- **Adopción y conversión**:
  - Tasa de adopción: mide la frecuencia con la que los usuarios seleccionan contenido recomendado.
  - Tasa de conversión: mide cuántos usuarios completan una acción específica (compra, vista, etc.) tras interactuar con una recomendación.

- **Ventas y ingresos**:
  - Incremento en ventas directas o indirectas atribuibles a recomendaciones.
  - Impacto en ingresos y márgenes de ganancia.

- **Efectos en la distribución de ventas**:
  - Analiza cambios en el patrón de ventas hacia productos menos populares (long tail) o más rentables.

- **Compromiso y comportamiento del usuario**:
  - Tiempo de permanencia en la plataforma.
  - Número de sesiones o interacciones tras recibir recomendaciones.

# Evaluación Sistemas de Recomendación


- La evaluación de los sistemas de recomendación es otro paso importante para evaluar la efectividad del método. 
- En la literatura podemos ver dos tipologías principales: evaluación offline y online.


##  Offline
- La evaluación offline es la que estamos acostumbrados en ML supervisado donde usamos un conjunto de entrenamiento y un conjunto de prueba; y aplicamos la métrica de evaluación apropiada. 
    
    Por ejemplo si se trata de etiquetas numéricas, como las calificaciones de 5 estrellas, la forma más común de validar un sistema de recomendación se basa en su valor de predicción, es decir, la capacidad de predecir las calificaciones del usuario. Las funciones estándar como el error cuadrático medio (RMSE), la precisión, las curvas ROC, etc...

    Sin embargo, Métricas como RMSE, precisión o recall no siempre correlacionan con el éxito comercial. Ejemplo: un algoritmo puede predecir bien si un usuario calificará positivamente un ítem, pero eso no garantiza que lo compre o lo use.
    


##  Online    
 
- La evaluación online es cuando utilizamos un conjunto de herramientas que nos permite observar las interacciones de los usuarios con el sistema. 
- La técnica en línea más común se llama prueba A-B y tiene la ventaja de permitir la evaluación del sistema al mismo tiempo que los usuarios aprenden, compran o juegan con el sistema de recomendación.

### Introducción a los Test A/B con Matemáticas

Los test A/B son una metodología experimental ampliamente utilizada para evaluar el impacto de cambios en productos, servicios o estrategias. Su objetivo es comparar dos versiones de una misma variable (A y B) y determinar cuál genera un mejor resultado en función de un criterio específico (como tasas de clics, conversiones o ingresos). 

Matemáticamente, los test A/B se basan en la inferencia estadística y el contraste de hipótesis. Aquí te explico los conceptos clave de su funcionamiento:



### 1. **Definición de Hipótesis**

El punto de partida es plantear las hipótesis nula (\(H_0\)) y alternativa (\(H_1\)):

- \(H_0\): No hay diferencia significativa entre las versiones A y B (la diferencia observada es por azar).
- \(H_1\): Existe una diferencia significativa entre las versiones A y B.

Por ejemplo, si queremos evaluar la tasa de clics (\(p\)), las hipótesis podrían ser:
\[
H_0: p_A = p_B \quad \text{y} \quad H_1: p_A \neq p_B.
\]



### 2. **Recopilación de Datos**

Se divide la población objetivo en dos grupos aleatorios:
- **Grupo A**: Recibe la versión A.
- **Grupo B**: Recibe la versión B.

Cada grupo produce un resultado medible, como la tasa de conversión (\(p_A\) y \(p_B\)), la media de ingresos (\(\mu_A\) y \(\mu_B\)), etc.



### 3. **Estadística del Test**

Para decidir entre \(H_0\) y \(H_1\), se calcula una estadística de prueba basada en los datos recolectados. Un ejemplo común es la prueba de diferencia de proporciones si trabajamos con tasas (\(p_A\) y \(p_B\)):

$$
z = \frac{(p_A - p_B)}{\sqrt{\hat{p}(1-\hat{p})\left(\frac{1}{n_A} + \frac{1}{n_B}\right)}},
$$
donde:
- $\hat{p}$ es la proporción combinada, $\hat{p} = \frac{x_A + x_B}{n_A + n_B}$,
- $n_A$ y $n_B$ son los tamaños de los grupos,
- $x_A$ y $x_B$ son los éxitos observados en cada grupo.

Si trabajamos con medias, el estadístico sería:


$$
t = \frac{(\bar{x}_A - \bar{x}_B)}{\sqrt{\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}}},
$$
donde \(\bar{x}_A\) y \(\bar{x}_B\) son las medias muestrales y \(s_A^2\), \(s_B^2\) son las varianzas muestrales.



### 4. **Decisión Estadística**

Se compara la estadística calculada (\(z\) o \(t\)) con un valor crítico de una distribución teórica (normal estándar o \(t\)-Student) para un nivel de significancia \(\alpha\) (por ejemplo, 0.05):

- Si el valor cae en la región crítica (\(p \leq \alpha\)), rechazamos \(H_0\).
- Si no, no podemos rechazar \(H_0\).



### 5. **Ejemplo Numérico**

Supongamos que queremos probar si una nueva versión de una página web mejora la tasa de clics. Los datos recolectados son:

- **Grupo A**: $x_A = 150$ clics de $n_A = 1000$ visitas ($p_A = 0.15$).
- **Grupo B**: $x_B = 180$ clics de $n_B = 1000$ visitas ($p_B = 0.18$).

Cálculo:
1. Proporción combinada: \(\hat{p} = \frac{150 + 180}{1000 + 1000} = 0.165\).
2. Estadística \(z\):
\[
z = \frac{(0.15 - 0.18)}{\sqrt{0.165 \cdot (1 - 0.165) \cdot \left(\frac{1}{1000} + \frac{1}{1000}\right)}} \approx -1.35.
\]
3. Comparamos \(z = -1.35\) con el valor crítico (\(\pm 1.96\) para \(\alpha = 0.05\)): no está en la región crítica, por lo que no rechazamos \(H_0\).



### 6. **Interpretación y Limitaciones**

En este ejemplo, no encontramos evidencia estadística suficiente para afirmar que la versión B mejora significativamente la tasa de clics. Sin embargo, esto no significa que B no sea mejor, solo que el efecto observado puede deberse al azar.

**Limitaciones**:
- Tamaño de muestra insuficiente (\(n\) pequeño).
- Resultados sesgados por aleatorización incorrecta.
- Mal ajuste del nivel de significancia \(\alpha\).

Los test A/B son una herramienta poderosa pero deben implementarse cuidadosamente para garantizar resultados confiables.

## Ejemplo en Python

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy.stats import ttest_ind

# Simulación de datos
np.random.seed(42)  # Para reproducibilidad
n_A = 10000  # Tamaño del grupo A
n_B = 10000  # Tamaño del grupo B

# Generar conversiones para cada grupo (1 = éxito, 0 = fracaso)
conversion_rate_A = 0.15
conversion_rate_B = 0.18

data_A = np.random.binomial(1, conversion_rate_A, n_A)
data_B = np.random.binomial(1, conversion_rate_B, n_B)

# Crear un DataFrame con los datos
df = pd.DataFrame({
    "success": np.concatenate([data_A, data_B]),
    "group": ["A"] * n_A + ["B"] * n_B
})

df
```

```python

# Agregar variables binarias para el grupo
df["group_A"] = (df["group"] == "A").astype(int)
df["group_B"] = (df["group"] == "B").astype(int)
df
```

```python
# Prueba t para diferencia de medias
t_stat, p_value = ttest_ind(data_A, data_B, equal_var=False)
```

### **Definición del \(p\)-value**
El \(p\)-value es la probabilidad de observar un resultado igual o más extremo que el obtenido en los datos, **bajo la suposición de que la hipótesis nula (\(H_0\)) es verdadera**. 

En términos prácticos:
- Si \(H_0\) es cierta (es decir, no hay diferencia real entre los grupos A y B), el \(p\)-value nos indica la probabilidad de que la diferencia observada (o una mayor) ocurra simplemente por azar.

```python

# Regresión para confirmar resultados
df["intercept"] = 1
model = sm.OLS(df["success"], df[["intercept", "group_B"]])
results = model.fit()

print(results.summary())
```
